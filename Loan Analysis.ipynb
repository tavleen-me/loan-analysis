# Analysis of Data

#### Now, in the data analysis step, we need to use some of the variables for the loan approval or rejection process and for this, we must aim to recognize the default predictors. For this purpose, firstly, the variables we have are the customer's details like age, occupation and employment details. Secondly, we have loan amount, interest rate and loan purpose and so on for determining loan features. At last, we have those variables, that are produced after the loan approval, which are, delinquent 2 years, revolving date and the next payment date and so on. These last variables describe customer charecteristics.

#### Since its the time of a loan application, we do not need customer characteristic variable as of now so we will drop that first for the analysis step.

#### For that, we will create a user defined variable where we will include all the customer characteristics attributes used after loan approval.

```python
# creating a customer characteristics variable
cust_char = ["delinq_2yrs", "earliest_cr_line", "inq_last_6mths", "open_acc", "pub_rec", "revol_bal", "revol_util", "total_acc", "out_prncp", 
             "out_prncp_inv",  "total_pymnt", "total_pymnt_inv", "total_rec_prncp", "total_rec_int", "total_rec_late_fee","recoveries",
             "collection_recovery_fee", "last_pymnt_d", "last_pymnt_amnt", "last_credit_pull_d", "application_type"]
```

```python
# eliminating cust_char variables
df = loan.drop(cust_char, axis=1)
```

#### Now for the loan_status attribute that gives the values for fully paid, curren ad charged off, we need to convert them into binary data (0 or 1) using astype() function for further analysis for further analysis.

#### The data 0 counts is for the default persons and the data 1 is for otherwise

```python
# converting loan_status to binary type
df['loan_status'] = df['loan_status'].astype('category')

# summarizing the values
df['loan_status'].value_counts()
```

#### Now, for the Current status, we are still in progress, so it will be eliminated as of now, but we do have Fully Paid and charged off, where we can assign them 0 or 1

```python
# processing the Fully Paid and Charged Off data
df['loan_status'] = df['loan_status'].apply(lambda x: 0 if x=='Fully Paid' else 1)
```

```python
# now converting the loan_status to integer
df['loan_status'] = df['loan_status'].apply(lambda x: pd.to_numeric(x))
```

```python
# getting the values for loan_status
df['loan_status'].value_counts()
```

```python
# checking the loan amount
sns.distplot(df['loan_amnt'])
plt.show()
```

#### Here, it shows the median loan amount is $10000 which shows the default rate of about 0.00006 

```python
# scattering the loan amount to high, medium, low and very high
def LoanAmount(n):
    if n < 5000:
        return 'low'
    elif n >= 5000 and n < 15000:
        return 'medium'
    elif n >= 15000 and n < 25000:
        return 'high'
    else:
        return 'very high'
        
df['loan_amnt'] = df['loan_amnt'].apply(lambda x: LoanAmount(x))
```

```python
df['loan_amnt'].value_counts()
```

```python
# compairing default rates across loan amount types
plt_var('loan_amnt')
```

##### Higher the loan amount higher the default rate

```python
# scattering the debt to income ratio into low, medium and high
def dti(n):
    if n <= 10:
        return 'low'
    elif n > 10 and n <= 20:
        return 'medium'
    else:
        return 'high'
    
df['dti'] = df['dti'].apply(lambda x: dti(x))
```

```python
# plotting the graph for the loan status across dti
plt_var('dti')
```

```python
# scattering the funded amount to low, medium and high
def FundedAmount(n):
    if n <= 5000:
        return 'low'
    elif n > 5000 and n <=15000:
        return 'medium'
    else:
        return 'high'
    
df['funded_amnt'] = df['funded_amnt'].apply(lambda x: FundedAmount(x))
```

```python
plt_var('funded_amnt')
```

##### Again higher the funded amount higher is the risk

```python
# Segmenting interest rate into higher, medium and lower
def IntRate(n):
    if n <= 10:
        return 'low'
    elif n > 10 and n <=15:
        return 'medium'
    else:
        return 'high'
    
    
df['int_rate'] = df['int_rate'].apply(lambda x: IntRate(x))
```

```python
# comparing default rate with interest rate
plt_var('int_rate')
```

```python
# now scattering installment into low medium and high
def Installment(n):
    if n <= 200:
        return 'low'
    elif n > 200 and n <=400:
        return 'medium'
    elif n > 400 and n <=600:
        return 'high'
    else:
        return 'very high'
    
df['installment'] = df['installment'].apply(lambda x: Installment(x))
```

```python
# plotting the graph and comparing default rates
plt_var('installment')
```

#### Higher the installment, higher would be the interest rate

```python
# comparing the annual income for default analysis
def AnnualIncome(n):
    if n <= 50000:
        return 'low'
    elif n > 50000 and n <=100000:
        return 'medium'
    elif n > 100000 and n <=150000:
        return 'high'
    else:
        return 'very high'

df['annual_inc'] = df['annual_inc'].apply(lambda x: AnnualIncome(x))
```

```python
# plotting the graph
plt_var('annual_inc')
```

#### Lower the income, lower will be the default rate

####  After comparing the default rate with various variables, we see that the grade, purpose of loan, annual income, interest rate etc. are the major default prdictors

# Analysing default rate data in asegmented manner

#### In this case, we will be analysing 4 major kinds of loans that lead to default analysis. Those loans are for the 'purpose variable' and are consolidation, credit card, home improvement and major purchase.

```python
# extracting 4 major loans from the purpose
purposes = ["credit_card","debt_consolidation","home_improvement","major_purchase"]
df = df[df['purpose'].isin(purposes)]
df['purpose'].value_counts()
```

```python
# graph plotting for number of loans by purpose
plt.figure(figsize=(10, 8))
sns.countplot(x=df['purpose']) # counting the number of plots
plt.show()
```

```python
# comparing default rates with purpose of loans and other variables like loan_status, grade, term, issue_d etc.
# checking it with the term
plt.figure(figsize=[8, 5])
sns.barplot(x='term', y="loan_status", hue='purpose', data=df)
plt.show()
```

```python
# writing a function to pass the categorial variable and then plotting the default rate 
def plt_seg(var):
    plt.figure(figsize=(8, 5))
    sns.barplot(x = var, y = 'loan_status', hue = 'purpose', data=df)
    plt.show()
plt_seg('term')
```
